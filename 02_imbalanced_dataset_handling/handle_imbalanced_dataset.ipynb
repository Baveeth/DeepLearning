###  Building an ANN Model in tensorflow/keras

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow_addons import losses
from sklearn.metrics import confusion_matrix , classification_report
```

```python
def ANN(X_train, y_train, X_test, y_test, loss, weights):
    model = keras.Sequential([
        keras.layers.Dense(26, input_dim=26, activation='relu'),
        keras.layers.Dense(15, activation='relu'),
        keras.layers.Dense(1, activation='sigmoid')
    ])

    model.compile(optimizer='adam', loss=loss, metrics=['accuracy'])
    
    if weights == -1:
        model.fit(X_train, y_train, epochs=100)
    else:
        model.fit(X_train, y_train, epochs=100, class_weight = weights)
    
    print(model.evaluate(X_test, y_test))
    
    y_preds = model.predict(X_test)
    y_preds = np.round(y_preds)
    
    print("Classification Report: \n", classification_report(y_test, y_preds))
    
    return y_preds
```

```python
y_preds = ANN(X_train, y_train, X_test, y_test, 'binary_crossentropy', -1)
```

### Method 1: Undersampling

```python
# Class Count
count_class_0, count_class_1 = df2.Churn.value_counts()

# Divide by Class
df_class_0 = df2[df2['Churn']==0]
df_class_1 = df2[df2['Churn']==1]
```

```python
df_class_0.shape
```

```python
# undersampling the no. of samples of majority class ("Churn=No") equals to minority class ("Churn=Yes")
df_class_0_under = df_class_0.sample(count_class_1)

#Concate Equal no. of samples (Churn, yes= 1869| Churn, No=1869) into one dataframe
df_test_under = pd.concat([df_class_0_under, df_class_1],axis=0)
```

```python
print("Random Under Sampling:")
print(df_test_under.Churn.value_counts())
```

```python
X = df_test_under.drop('Churn', axis='columns')
y = df_test_under['Churn']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5, stratify=y)
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

```python
y_train.shape
```

```python
y_preds = ANN(X_train, y_train, X_test, y_test, 'binary_crossentropy', -1)
```

### Method 3: SMOTE (Synthetic Minority Oversampling Technique)

```python
X = df2.drop('Churn', axis='columns')
y = df2['Churn']
```

```python
#pip install imbalanced-learn (installing imbalanced learn library)

from imblearn.over_sampling import SMOTE

smote = SMOTE(sampling_strategy='minority')
X_sm, y_sm = smote.fit_resample(X, y)

y_sm.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_sm, y_sm, test_size=0.2, random_state=5, stratify=y_sm)
```

```python
y_preds =ANN(X_train, y_train, X_test, y_test,'binary_crossentropy', -1)
```
